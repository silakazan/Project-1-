```python
MTA_data.drop_duplicates(subset=None, keep="first", inplace=True)
MTA_data.describe()
```

```python
MTA_data["TURNSTILE"] = MTA_data["C/A"]+"-"+MTA_data["UNIT"]+"-"+MTA_data["SCP"]

MTA_data.head(10)
```

```python
# Creating DATETIME as datetime64 column that concatenate "DATE" and "TIME"

MTA_data['DATETIME'] = MTA_data["DATE"]+ " " + MTA_data["TIME"]

MTA_data['DATETIME'] = pd.to_datetime(MTA_data['DATETIME'], format='%m/%d/%Y %H:%M:%S')

MTA_data.info()
```

```python
#Create a column that called "NAMEOFDAY"
MTA_data["DAY"]= MTA_data["DATETIME"].dt.day_name()

#Checking sample 5 rows
MTA_data.sample(5)
```

```python
# Working with only neccessary columns

MTA_data = MTA_data[["STATION","TURNSTILE","DATE","TIME","DATETIME","DAY","ENTRIES","EXITS"]]
MTA_data.head(10)
```

```python
#Finding Daily traffic(entries+exits) based on turnstile
MTA_data_entries = MTA_data.groupby(["STATION","TURNSTILE","DATE","DAY"]).ENTRIES.max() - MTA_data.groupby(["STATION","TURNSTILE","DATE","DAY"]).ENTRIES.min()

MTA_data_exits = MTA_data.groupby(["STATION","TURNSTILE","DATE","DAY"]).EXITS.max() - MTA_data.groupby(["STATION","TURNSTILE","DATE","DAY"]).EXITS.min()

MTA_data_entries = MTA_data_entries.reset_index()

MTA_data_exits = MTA_data_exits.reset_index()

MTA_data_entries_exits = pd.merge(MTA_data_entries, MTA_data_exits, how="outer")

MTA_data_entries_exits["TRAFFIC"] = MTA_data_entries_exits["ENTRIES"] + MTA_data_entries_exits["EXITS"]

MTA_data_entries_exits.sample(10)
```

```python
#Boxenplot for daily traffic

ax = sns.boxenplot(x='TRAFFIC', data=MTA_data_entries_exits,palette="bright").set(xlabel='Daily Traffic',title="Daily Traffic Based On Turnstile")
```

```python
MTA_data_traffic_under_percentileX = MTA_data_entries_exits[MTA_data_entries_exits["TRAFFIC"]<(MTA_data_entries_exits.TRAFFIC.quantile(0.90))]
ax=sns.distplot(MTA_data_traffic_under_percentileX["TRAFFIC"], bins=16, color="PURPLE")
```

```python
MTA_data_traffic_under_percentileX = MTA_data_entries_exits[MTA_data_entries_exits["TRAFFIC"]<(MTA_data_entries_exits.TRAFFIC.quantile(0.90))]
ax=sns.histplot(MTA_data_traffic_under_percentileX["TRAFFIC"], bins=16, color="PURPLE")
```

```python
# Finding top 4 busiest stations based on traffic

MTA_data_top = MTA_data_entries_exits.groupby("STATION").sum().sort_values("TRAFFIC",ascending=False).head(4)
MTA_data_top = MTA_data_top.reset_index()
MTA_data_top
```

```python
#Barplot for Top 4 Busiest Stations

plt.figure(figsize=(10,5),dpi=200)
plt.style.use('default')
ax = sns.barplot(x="STATION", y="TRAFFIC", data=MTA_data_top)
plt.title("Top 4 Busiest Stations", weight="bold", c="purple",fontsize=15)
plt.xlabel("",weight="bold",c="k")
plt.xticks(rotation=0, weight="bold")
plt.ylabel("Total Traffic)",weight= "bold",c="k")
plt.ylim(0,30000000)
plt.yticks(weight="bold")
plt.show()
```

```python
# Finding overall busiest day based on traffic

MTA_data_busiest_day = MTA_data_entries_exits.groupby("DAY").sum().sort_values("TRAFFIC",ascending=False)
MTA_day_busiest_day = MTA_data_busiest_day.reset_index()
MTA_data_busiest_day

```

```python
plt.figure(figsize=(10,5),dpi=200)
plt.style.use('default')
ax = sns.barplot(x="DAY", y="TRAFFIC", data=MTA_data_busiest_day)
plt.title("Top 4 Busiest Stations", weight="bold", c="purple",fontsize=15)
plt.xlabel("",weight="bold",c="k")
plt.xticks(rotation=0, weight="bold")
plt.ylabel("Total Traffic)",weight= "bold",c="k")
plt.ylim(0,30000000)
plt.yticks(weight="bold")
plt.show()
```
