```python
df["TURNSTILE"] = df["C/A"]+"-"+df["UNIT"]+"-"+df["SCP"]

df.head(10)
```

```python
df['DATETIME'] = df["DATE"]+ " " + df["TIME"]
df['DATETIME'] = pd.to_datetime(df['DATETIME'], format='%m/%d/%Y %H:%M:%S')
df.info()
```

```python
df["NAMEOFDAY"]= df["DATETIME"].dt.day_name()
```

```python
df = df[["STATION","TURNSTILE","DATE","TIME","DATETIME","NAMEOFDAY","ENTRIES","EXITS"]]
df.head(10)
```

```python
ax = sns.scatterplot(x="DATE", y="ENTRIES", data=df)
plt.xticks("")
plt.show()
```

```python
df_entries = df.groupby(["STATION","TURNSTILE","DATE","NAMEOFDAY"]).ENTRIES.max() - df.groupby(["STATION","TURNSTILE","DATE","NAMEOFDAY"]).ENTRIES.min()

df_exits = df.groupby(["STATION","TURNSTILE","DATE","NAMEOFDAY"]).EXITS.max() - df.groupby(["STATION","TURNSTILE","DATE","NAMEOFDAY"]).EXITS.min()

df_entries = df_entries.reset_index()

df_exits = df_exits.reset_index()

df_entries_exits = pd.merge(df_entries, df_exits, how="outer")

df_entries_exits["TRAFFIC"] = df_entries_exits["ENTRIES"] + df_entries_exits["EXITS"]

df_entries_exits.sample(10)
```

```python
ax = sns.boxenplot(x='TRAFFIC', data=df_entries_exits,palette="bright").set(xlabel='Daily Traffic',title="Daily Traffic Based On Turnstile")
```
